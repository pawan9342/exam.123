```python
# Task 5: Collect strings, then display them in lower case alphabetical order

strings = []

while True:
    user_input = input("Enter a string (press Enter to finish): ")
    if user_input == "":
        break
    strings.append(user_input)

# Convert to lower case and sort alphabetically
strings = [s.lower() for s in strings]
strings.sort()

print("Strings in alphabetical order:")
for s in strings:
    print(s)
```
